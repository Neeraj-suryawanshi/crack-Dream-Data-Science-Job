**Question:1** What is the output of the following Python code?
```python

x = 10

y = 20

if x == 10:

    x = x + y

print(x)

```
**Options:**
```python
1) 10

2) 20

3) 30

4) None of the above

```
**Answer:** 30

**Explanation:**
In this code, x is first defined as 10. Then y is defined as 20. In the if statement, x is reassigned to x + y, which evaluates to 10 + 20 = 30. Finally, x is printed, resulting in 30.

**Question:7** What does the range() function do in a for loop?

**Options:**
```python

1) Determines the number of times the loop will run

2) Specifies the values that the loop variable will take on

3) Determines the starting and ending values of the loop

4) All of the above

```
**Answer:** All of the above.

The range() function can take one, two, or three arguments, and the arguments determine the starting value, ending value, and step size of the range.

# Quiz 1

### 1. What is multithreading in Python?

- A. A process of running multiple threads simultaneously within a single process.
- B. A process of running multiple processes simultaneously within a single thread.
- C. A process of running multiple processes simultaneously within multiple threads.
- D. A process of running a single process within multiple threads.

**Answer:** A

**Explanation:** Multithreading in Python is the process of running multiple threads simultaneously within a single process.

---

### 2. What is the purpose of multithreading in Python?

- A. To improve the performance of a program by using multiple CPUs or CPU cores.
- B. To improve the memory utilization of a program by using multiple CPUs or CPU cores.
- C. To improve the overall efficiency of a program by using multiple CPUs or CPU cores.
- D. To improve the input/output operations of a program by using multiple CPUs or CPU cores.

**Answer:** A

**Explanation:** The primary purpose of multithreading in Python is to improve the performance of a program by using multiple CPUs or CPU cores.

---

### 3. Which module is used to create and manage threads in Python?

- A. threading
- B. multiprocessing
- C. concurrent.futures
- D. queue

**Answer:** A

**Explanation:** The threading module is used to create and manage threads in Python.

---

### 4. What is a mutex in multithreading?

- A. A type of thread that performs a specific task.
- B. A type of thread that runs in the background to handle incoming requests.
- C. A synchronization primitive used to protect shared resources from being accessed by multiple threads at the same time.
- D. A built-in function in Python that is used to release the Global Interpreter Lock.

**Answer:** C

**Explanation:** A mutex is a synchronization primitive used to protect shared resources from being accessed by multiple threads at the same time.

---

### 5. Which method is used to start a new thread in Python?

- A. start()
- B. run()
- C. create()
- D. begin()

**Answer:** A

**Explanation:** The start() method is used to start a new thread in Python.

---

### 6. What is the Global Interpreter Lock (GIL) in Python?

- A. A lock used to protect shared resources from being accessed by multiple threads at the same time.
- B. A mechanism used by Python to ensure that only one thread executes Python bytecode at a time.
- C. A thread that runs in the background to handle incoming requests.
- D. A type of thread that performs a specific task.

**Answer:** B

**Explanation:** The Global Interpreter Lock (GIL) is a mechanism used by Python to ensure that only one thread executes Python bytecode at a time.

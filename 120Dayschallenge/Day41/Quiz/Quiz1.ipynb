1. **What is the most common airline in the Flight Price dataset?**
   - Delta
   - American Airlines
   - United
   - Southwest

   **Answer:** Southwest

   **Explanation:** To identify the most common airline, we can use `value_counts()` on the 'Airline' column, which returns the count of each unique value. Southwest has the highest count, indicating it is the most common airline.

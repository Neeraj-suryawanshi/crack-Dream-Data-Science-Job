# Quiz 1

### 1. Which type of probability distribution is used to model the number of successes in a fixed number of independent trials?

- Binomial Distribution
- Poisson Distribution
- Normal Distribution
- Exponential Distribution

**Answer:** Binomial Distribution

**Explanation:** Binomial distribution is used to model the number of successes in a fixed number of independent trials, where each trial has only two possible outcomes (success or failure).

---

### 2. Which probability distribution is used to model the number of events occurring in a fixed interval of time or space?

- Binomial Distribution
- Poisson Distribution
- Normal Distribution
- Exponential Distribution

**Answer:** Poisson Distribution

**Explanation:** Poisson distribution is used to model the number of events occurring in a fixed interval of time or space, such as the number of cars passing through a toll booth in an hour or the number of calls received by a call center in a day.

---

### 3. Which probability distribution is symmetric and bell-shaped?

- Binomial Distribution
- Poisson Distribution
- Normal Distribution
- Exponential Distribution

**Answer:** Normal Distribution

**Explanation:** Normal distribution is symmetric and bell-shaped. It is widely used to model many natural phenomena, such as the heights and weights of people and the errors in measurements.

---

### 4. In a binomial distribution, what is the formula for calculating the probability of k successes in n trials?

- P(k) = nCk * p^k * (1-p)^(n-k)
- P(k) = e^(-λ) * λ^k / k!
- P(k) = μ + σZ
- P(k) = λ^k * e^(-λ) / k!

**Answer:** P(k) = nCk * p^k * (1-p)^(n-k)

**Explanation:** This is the formula for the probability of k successes in n trials, where p is the probability of success and (1-p) is the probability of failure.

---

### 5. In a Poisson distribution, what is the formula for calculating the probability of k events occurring in a fixed time interval?

- P(k) = nCk * p^k * (1-p)^(n-k)
- P(k) = e^(-λ) * λ^k / k!
- P(k) = μ + σZ
- P(k) = λ^k * e^(-λ) / k!

**Answer:** P(k) = λ^k * e^(-λ) / k!

**Explanation:** This is the formula for the probability of k events occurring in a fixed time interval, where λ is the expected number of events in that time interval.

---

### 6. In a normal distribution, what is the percentage of data that falls within one standard deviation of the mean?

- 68%
- 95%
- 99.7%
- 100%

**Answer:** 68%

**Explanation:** In a normal distribution, approximately 68% of data falls within one standard deviation of the mean.

---

### 7. In a binomial distribution, what happens to the shape of the distribution as the number of trials increases?

- It becomes wider
- It becomes narrower
- It becomes taller
- It becomes flatter

**Answer:** It becomes narrower

**Explanation:** As the number of trials in a binomial distribution increases, the shape of the distribution becomes narrower and more concentrated around the expected value.

---

### 8. In a Poisson distribution, what happens to the shape of the distribution as the mean increases?

- It becomes wider
- It becomes narrower
- It becomes taller
- It becomes flatter

**Answer:** It becomes taller

**Explanation:** As the mean of a Poisson distribution increases, the shape of the distribution becomes taller.

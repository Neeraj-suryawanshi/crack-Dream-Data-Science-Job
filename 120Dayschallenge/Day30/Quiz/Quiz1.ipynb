# Quiz 1

### 1. What is a hypothesis in statistics?

- A guess about the outcome of an experiment
- A proven fact
- A random number generator
- A programming language

**Answer:** A guess about the outcome of an experiment

**Explanation:** A hypothesis is an assumption or prediction about the outcome of an experiment or study.

---

### 2. What is the mechanism of hypothesis testing?

- Analyzing data
- Conducting experiments
- Formulating a null and alternative hypothesis
- All of the above

**Answer:** Formulating a null hypothesis and an alternative hypothesis.

**Explanation:** The mechanism of hypothesis testing involves formulating a null hypothesis (which assumes there is no effect or relationship between variables) and an alternative hypothesis (which assumes there is an effect or relationship), collecting data, analyzing the data, and determining whether the null hypothesis can be rejected or not.

---

### 3. What is the significance level commonly used in hypothesis testing?

- 0.01
- 0.05
- 0.10
- 0.25

**Answer:** 0.05

**Explanation:** The significance level commonly used in hypothesis testing is 0.05, indicating a 5% chance of rejecting the null hypothesis when it is true.

---

### 4. Which of the following is a statement that is being tested in hypothesis testing?

- Alternative fact
- Null hypothesis
- Alternative hypothesis
- Alternative reality

**Answer:** Null hypothesis

**Explanation:** In hypothesis testing, the null hypothesis is a statement that is being tested against the alternative hypothesis to determine whether there is evidence to reject the null hypothesis.

---

### 5. What is estimation in statistics?

- A method used to test hypotheses about a population parameter
- The process of drawing conclusions about a population based on sample data
- The process of calculating the mean and standard deviation of a population
- The process of collecting data from a population

**Answer:** The process of drawing conclusions about a population based on sample data

**Explanation:** Estimation in statistics is the process of drawing conclusions about a population based on sample data. This involves using sample statistics to make inferences about population parameters.

---

### 6. What is the mechanism of hypothesis testing?

- Collecting data, conducting experiments, and analyzing data
- Formulating a null hypothesis and an alternative hypothesis, collecting data, and analyzing data
- Calculating the mean and standard deviation of a population
- Drawing conclusions about a population based on sample data

**Answer:** Formulating a null hypothesis and an alternative hypothesis, collecting data, and analyzing data

**Explanation:** The mechanism of hypothesis testing involves formulating a null hypothesis and an alternative hypothesis, collecting data, and analyzing data to determine whether the null hypothesis can be rejected or not.

### 7. What is a p-value?

- The probability of rejecting the null hypothesis
- The probability of observing a test statistic as extreme as, or more extreme than, the one observed, assuming the null hypothesis is true
- The probability of observing a test statistic as extreme as, or more extreme than, the one observed, assuming the alternative hypothesis is true
- The probability of observing a test statistic as extreme as, or more extreme than, the one observed, without assuming any hypothesis is true

**Answer:** The probability of observing a test statistic as extreme as, or more extreme than, the one observed, assuming the null hypothesis is true

**Explanation:** A p-value is the probability of observing a test statistic as extreme as, or more extreme than, the one observed, assuming the null hypothesis is true. It is used to determine the statistical significance of the results.

---

### 8. What is the student t distribution?

- A probability distribution used to test hypotheses about the mean of a population when the sample size is small
- A probability distribution used to test hypotheses about the mean of a population when the sample size is large
- A measure of central tendency
- A measure of variability

**Answer:** A probability distribution used to test hypotheses about the mean of a population when the sample size is small

**Explanation:** The student t distribution is used when the sample size is small and the population standard deviation is unknown and must be estimated from the sample data.

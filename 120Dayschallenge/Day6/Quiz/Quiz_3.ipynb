**Question:4** What will be the output of the following code:
```python
sum = 0

for i in range(1,100):

    sum += i

    if sum > 100:

        break

print(sum)
```
**Options:**
- A number equal to 100
- A number less than 100
- A number greater than 100
- None

**Answer:** A number greater than 100 

**Explanation:** The loop will start summing up the numbers from 1 to 100. In each iteration, the sum of the numbers will increase. When the sum value will be greater than 100 the loop will be terminated. So the Sum will be a number greater than 100.

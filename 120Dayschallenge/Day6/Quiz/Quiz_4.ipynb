**Question:5** What will be the output of the following code:
```python
list_out=[]

for i in range(0,51,10):

    list_out.append(str(i))

print(list_out)

```
**Options:**
```python
 [0, 10, 20, 30, 40]

 ['0', '10', '20', '30', '40']

 ['0', '10', '20', '30', '40', '50']

 [0, 10, 20, 30, 40, 50]
```
**Answer:** 
 ['0', '10', '20', '30', '40', '50']

**Explanation:** The for loop runs over a range of 1 to 50 with a step size of 10. In each iteration, the value of i is converted to a string datatype and is appended to the list_out list.

**Question 9:**
Which two methods are used to join the two sets?

1. + and join()
2. union() and update()
3. append() and extend()
4. None

Options:
- 1, and 2 are correct.
- 1, and 3 are correct.

**Question 10:**
Which of the following methods are used for tuples and sets?

Tuples: count() and index() Tuples: append() and extend() Sets: add(), pop() and union() Sets: union(), update(), and index()

Options:
- 1 and 2 are correct.
- 1 and 3 are correct.
- 1, 2, 3, and 4 are correct.
- 2, and 3 are correct.

Answer: 2, and 3 are correct.

Explanation: `count()` and `index()` are the only methods available for tuples. This is because tuples are immutable. `add()`, `pop()`, `union()`, and `update()` are methods for the set, but `index()` is used for lists and tuples.

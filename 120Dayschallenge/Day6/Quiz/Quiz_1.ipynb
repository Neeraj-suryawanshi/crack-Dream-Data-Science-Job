1. Which of the following are the type-casting methods in Python Programming language?
   - 1. int()
   - 2. type()
   - 3. float()
   - 4. str()
   
   Options:
   - 1 and 2
   - 1, 3 and 4
   - 1, 2, and 3 
   - 1, 2, 3 and 4 

   **Answer**: 1, 3 and 4
   **Explanation**: type() is an inbuilt python method used to check the type of any variable in python. int(), str() and float() are basic type casting methods in python. Apart from this list(), tuple(), dict() are also some examples of type-casting methods.

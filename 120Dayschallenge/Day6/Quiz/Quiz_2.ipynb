**Question 1**

* **Which of the following are the type-casting methods in Python Programming language?**

**Options:**

1. `int()`
2. `type()`
3. `float()`
4. `str()`

**Answer:**
1. `int()`
2. `type()`
3. `float()`
4. `str()`

**Explanation:**
`type()` is used to check the type of any variable. `int()`, `str()`, and `float()` are basic type casting methods in Python.

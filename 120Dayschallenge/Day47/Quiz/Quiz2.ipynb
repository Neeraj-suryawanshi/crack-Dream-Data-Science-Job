# Quiz 2

### 1. Which of the following is a technique for tuning hyperparameters that exhaustively searches a specified subset of the hyperparameter space?

- Randomize search cv
- Data leakage
- Confusion matrix
- Grid search cv

**Answer:** Grid search cv

**Explanation:** Grid search cv is a technique that exhaustively searches a specified subset of the hyperparameter space.

### 2. Which of the following is a technique for tuning hyperparameters that randomly samples a specified subset of the hyperparameter space?

- Data leakage
- Confusion matrix
- Randomize search cv
- Grid search cv

**Answer:** Randomize search cv

**Explanation:** Randomize search cv is a technique that randomly samples a specified subset of the hyperparameter space.

### 3. What is data leakage?

- When data is accidentally or intentionally leaked to the public
- When the test data is used to train the model
- When the training data contains information that would not be available in the real world
- None of the above

**Answer:** When the training data contains information that would not be available in the real world

**Explanation:** Data leakage occurs when the training data contains information that would not be available in the real world, leading to overfitting and inaccurate results.

### 4. Which of the following is a tool used to evaluate the performance of a classification model?

- Grid search cv
- Randomize search cv
- Data leakage
- Confusion matrix

**Answer:** Confusion matrix

**Explanation:** A confusion matrix is a tool used to evaluate the performance of a classification model.

### 5. What is the precision of a model?

- The number of true positives divided by the number of false positives
- The number of true positives divided by the sum of true positives and false negatives
- The number of true positives divided by the sum of true positives and false positives
- The number of true negatives divided by the sum of true negatives and false positives

**Answer:** The number of true positives divided by the sum of true positives and false positives

**Explanation:** Precision measures the proportion of true positives out of all predicted positives.

### 6. What is the recall of a model?

- The number of true positives divided by the sum of true positives and false negatives
- The number of true positives divided by the sum of true positives and false positives
- The number of true positives divided by the number of false positives
- The number of true negatives divided by the sum of true negatives and false positives

**Answer:** The number of true positives divided by the sum of true positives and false negatives

**Explanation:** Recall measures the proportion of true positives out of all actual positives.

### 7. What is the F1 score of a model?

- The weighted average of precision and recall
- The harmonic mean of precision and recall
- The arithmetic mean of precision and recall
- None of the above

**Answer:** The harmonic mean of precision and recall

**Explanation:** The F1 score is a weighted average of precision and recall, with more emphasis on the smaller value.

### 8. Which of the following hyperparameters can be tuned using grid search cv for a logistic regression classifier?

- C
- Penalty
- L1 ratio
- n_neighbors

**Answer:** L1 ratio

**Explanation:** The hyperparameters that can be tuned using grid search cv for a logistic regression classifier are C and the choice of penalty ('l1' or 'l2').

4. What is the best way to iterate over the rows of a Pandas DataFrame?

   - Using a for loop to iterate through the rows by index.
   - Using the apply() method to apply a function to each row.
   - Using the iterrows() method.
   - Using the itertuples() method.

   **Answer:** Using the itertuples() method.

   **Explanation:** The itertuples() method is generally considered more efficient than iterrows(). It returns an iterator of namedtuples, providing a good balance between performance and ease of use when iterating over the rows of a Pandas DataFrame. The apply() method is also a viable option, especially when you want to apply a function to each row.


5. What is the difference between the iterrows() method and the itertuples() method for iterating over a Pandas DataFrame?
   
   - The iterrows() method is faster but yields a Series, while the itertuples() method is slower but yields a named tuple.
   - The iterrows() method yields a Series, while the itertuples() method yields a DataFrame.
   - The iterrows() method yields a Series, while the itertuples() method yields a named tuple.
   - The iterrows() method yields a DataFrame, while the itertuples() method yields a Series.

   **Answer:** The iterrows() method yields a Series, while the itertuples() method yields a named tuple.

   **Explanation:** The iterrows() method returns an iterator that yields index and row data for each row as a Series object. On the other hand, the itertuples() method returns an iterator that yields a named tuple for each row, where the values are accessed by field names. While itertuples() is generally faster than iterrows(), iterrows() yields a Series, and itertuples() yields a more memory-efficient named tuple.

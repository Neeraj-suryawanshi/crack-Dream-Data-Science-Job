# Quiz 1

1. What is the F-test used for?
   - To test the equality of variances in two populations
   - To test the equality of means in two populations
   - To test the correlation between two variables
   - To test the goodness-of-fit of a model

   **Answer:** To test the equality of variances in two populations.

   **Explanation:** The F-test (variance ratio test) is used to determine if the variances of two populations are equal or not.

2. What are the possible values of F?
   - Only positive values
   - Only negative values
   - Both positive and negative values
   - Zero and positive values

   **Answer:** Only positive values.

   **Explanation:** F is always a positive number or zero. It is the ratio of two variances, which cannot be negative.

3. Which of the following is NOT a parameter for the F-distribution?
   - Degree of freedom for the numerator
   - Degree of freedom for the denominator
   - Mean
   - Standard deviation

   **Answer:** Mean.

   **Explanation:** The F-distribution depends on degrees of freedom for the numerator and denominator, not on the mean.

4. Which library in Python can be used to calculate the F-distribution?
   - NumPy
   - SciPy
   - Pandas
   - Matplotlib

   **Answer:** SciPy.

   **Explanation:** The SciPy library in Python provides the f-distribution module to calculate the F-distribution.

5. What is the significance level for the F-test?
   - The probability of rejecting the null hypothesis when it is true
   - The probability of accepting the null hypothesis when it is false
   - The probability of obtaining a test statistic as extreme as or more extreme than the observed value, assuming the null hypothesis is true
   - The probability of obtaining a test statistic as extreme as or more extreme than the observed value, assuming the alternative hypothesis is true

   **Answer:** The probability of obtaining a test statistic as extreme as or more extreme than the observed value, assuming the null hypothesis is true.

   **Explanation:** The significance level sets the threshold for rejecting the null hypothesis based on the p-value.

6. Two samples have variances 4 and 9. What is the value of F?
   - 0.44
   - 1.125
   - 1.5
   - 2.25

   **Answer:** 2.25.

   **Explanation:** F = 9/4 = 2.25.

7. For an F-distribution with df1=6 and df2=9, what is the value of F for a cumulative probability of 0.95?
   - 3.43
   - 3.69
   - 3.89
   - 4.23

   **Answer:** 3.69.

   **Explanation:** Using the f.ppf() function from SciPy library, f.ppf(0.95, 6, 9) = 3.69.

8. A researcher wants to test if the variances of two samples are equal. She collects two samples of sizes 20 and 30 with sample variances 12 and 18, respectively. What is the calculated value of F?
   - 0.67
   - 0.75
   - 1.25
   - 1.5

   **Answer:** 1.5.

   **Explanation:** F = s1^2/s2^2 = 12/18 = 1.5.

9. Given an F-distribution with df1=8 and df2=12, what is the probability that F is greater than 3.50?
   - 0.026
   - 0.041
   - 0.074
   - 0.096

   **Answer:** 0.074.

   **Explanation:** Using the f.sf() function from SciPy library, f.sf(3.50, 8, 12) = 0.074.

10. An experiment is conducted to compare the variances of two groups of data. The F-value is calculated to be 1.75, with degrees of freedom for the numerator and denominator being 6 and 8, respectively. What is the p-value for this F-value?
    - 0.105
    - 0.155
    - 0.205
    - 0.255

    **Answer:** 0.155.

    **Explanation:** Using the f.cdf() function from SciPy library, f.cdf(1.75, 6, 8) = 0.845, so p = 1 - 0.845 = 0.155.

#### Q1. How can you create a Bokeh plot using Python code?

```python
"""
Ans: Bokeh produces visuals for web browsers in HTMl. 
Can be created as shown below:
"""

from bokeh.plotting import figure, output_file, show
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]
output_file("bokeh_plot.html")

p = figure(title="Bokeh Line Plot", x_axis_label='X-axis', y_axis_label='Y-axis')
p.line(x, y, legend_label="Line", line_width=2)
show(p)
```

#### Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

```python
"""
Ans: A glyph is a graphical shape used to represent data visually. Glyphs can be used to create a wide range of plot types, from simple scatter plots to complex visualizations like heatmaps and network graphs.

Here's an example of how we can add glyphs:
"""
from bokeh.plotting import figure, output_file, show
from bokeh.models import ColumnDataSource


x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]
colors = ['red', 'green', 'blue', 'orange', 'purple']

output_file("bokeh_glyphs.html")
source = ColumnDataSource(data=dict(x=x, y=y, color=colors))

p = figure(title="Bokeh Glyphs Plot", x_axis_label='X-axis', y_axis_label='Y-axis')
p.circle('x', 'y', color='color', size=10, source=source, legend_label="Circle")
p.square('x', 'y', color='color', size=15, source=source, legend_label="Square")

show(p)
```

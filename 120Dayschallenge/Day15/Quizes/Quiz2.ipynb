# Quiz 2


1. **What is the name of the official MongoDB driver for Python?**
   - MongoPython
   - PyMongoDB
   - MongoDB-Py
   - PyMongo

   **Answer:** PyMongo

   **Explanation:** PyMongo is the official MongoDB driver for Python. It is a Python library that allows you to interact with MongoDB from Python, providing a simple and convenient way to connect to MongoDB and perform operations.

2. **How do you connect to MongoDB using PyMongo in Python?**
   - By using the connect() method
   - By using the PyMongo() method
   - By using the MongoClient() method
   - By using the MongoDB() method

   **Answer:** By using the MongoClient() method

   **Explanation:** To connect to MongoDB using PyMongo in Python, you need to use the MongoClient() method. It takes a connection string as an argument and returns a MongoClient instance for interacting with the database.

3. **How do you access a collection in MongoDB using PyMongo in Python?**
   - By using the collection() method
   - By using the get_collection() method
   - By using the get_database() method
   - By using the database[collection] syntax

   **Answer:** By using the database[collection] syntax

   **Explanation:** To access a collection in MongoDB using PyMongo in Python, you can use the database[collection] syntax, allowing you to perform CRUD operations on the collection.

4. **How do you insert a document into a collection in MongoDB using PyMongo in Python?**
   - By using the insert_one() method
   - By using the insert() method
   - By using the save() method
   - By using the add() method

   **Answer:** By using the insert_one() method

   **Explanation:** To insert a document into a collection in MongoDB using PyMongo in Python, use the insert_one() method, providing the document as an argument.

5. **How do you query documents from a collection in MongoDB using PyMongo in Python?**
   - By using the find() method
   - By using the query() method
   - By using the get() method
   - By using the select() method

   **Answer:** By using the find() method

   **Explanation:** To query documents from a collection in MongoDB using PyMongo in Python, use the find() method, which returns a cursor for iterating over matching documents.

6. **What is MongoDB?**
   - A relational database management system
   - A document-oriented database management system
   - A graph database management system
   - A key-value store database management system

   **Answer:** A document-oriented database management system

   **Explanation:** MongoDB is a document-oriented database management system that stores data as documents in BSON format, providing flexibility in data modeling.

7. **What is a collection in MongoDB?**
   - A database
   - A table
   - A set of documents
   - A set of rows

   **Answer:** A set of documents

   **Explanation:** In MongoDB, a collection is a set of documents, equivalent to a table in relational databases.

8. **What is a document in MongoDB?**
   - A row
   - A tuple
   - An instance of data
   - A JSON-like structure

   **Answer:** A JSON-like structure

   **Explanation:** In MongoDB, a document is a JSON-like structure representing an instance of data, with fields of various data types.

9. **What is the default storage engine for MongoDB?**
   - RocksDB
   - WiredTiger
   - InnoDB
   - MyISAM

   **Answer:** WiredTiger

   **Explanation:** WiredTiger is the default storage engine for MongoDB, offering high performance, scalability, and features like document-level locking.

10. **What is the purpose of the MongoDB aggregation framework?**
    - To perform complex data processing on collections
    - To enforce data consistency and integrity
    - To manage transactions and rollbacks
    - To define indexing strategies

    **Answer:** To perform complex data processing on collections

    **Explanation:** The MongoDB aggregation framework is designed for complex data processing on collections, offering operations like grouping, filtering, and transformation through a pipeline-based interface.

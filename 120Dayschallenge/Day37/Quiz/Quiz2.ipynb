# Quiz 2


1. **What is the purpose of Min-Max Scaling?**
   - To standardize data using mean and standard deviation
   - To normalize data into a range of [0,1]
   - To remove outliers from the data
   - To perform feature extraction

   **Answer:** To normalize data into a range of [0,1]

   **Explanation:** Min-Max Scaling transforms the data into a specified range, commonly [0,1], making it suitable for analysis.

2. **What is Feature Extraction?**
   - The process of selecting the most important features from a dataset
   - The process of transforming data into a different representation
   - The process of combining multiple datasets into a single dataset
   - The process of removing outliers from a dataset

   **Answer:** The process of transforming data into a different representation

   **Explanation:** Feature Extraction involves transforming data into a more suitable representation for analysis, which can include selecting important features or creating new ones.

3. **What is PCA (Principal Component Analysis)?**
   - A feature extraction technique that transforms data into a lower-dimensional space
   - A classification algorithm used for image recognition
   - A clustering algorithm used for unsupervised learning
   - An outlier detection technique used for anomaly detection

   **Answer:** A feature extraction technique that transforms data into a lower-dimensional space

   **Explanation:** PCA is a technique for dimensionality reduction, finding principal components that capture the most variance in the data.

4. **What is the purpose of PCA?**
   - To standardize data using mean and standard deviation
   - To normalize data into a range of [0,1]
   - To reduce the dimensionality of the data
   - To remove outliers from the data

   **Answer:** To reduce the dimensionality of the data

   **Explanation:** PCA aims to find a lower-dimensional representation of the data while retaining the most important information.

5. **What is the range of values after performing Min-Max Scaling?**
   - [-1, 1]
   - [0, 1]
   - [1, 10]
   - [-10, 10]

   **Answer:** [0, 1]

   **Explanation:** Min-Max Scaling transforms the data into a range, commonly [0,1], by normalizing it.

6. **What is the L2 norm of a unit vector?**
   - 1
   - 0
   - -1
   - Infinity

   **Answer:** 1

   **Explanation:** The L2 norm of a unit vector (length 1) is 1.

7. **Which of the following is a feature extraction technique?**
   - Min-Max Scaling
   - Unit Vector
   - PCA
   - K-Means Clustering

   **Answer:** PCA

   **Explanation:** PCA is a feature extraction technique that transforms data into a lower-dimensional space.

8. **A data scientist is working with a dataset of stock prices that vary widely in magnitude. Which of the following data preprocessing techniques should they use to ensure that all features have equal importance in the analysis?**
   - Min-Max Scaling
   - Unit Vector
   - Feature Extraction
   - PCA

   **Answer:** Min-Max Scaling

   **Explanation:** Min-Max Scaling ensures that features have equal importance by normalizing them into a specified range.

9. **An analyst is working with a dataset of customer demographics and wants to reduce the number of features to improve the efficiency of the analysis. Which of the following techniques should they use?**
   - Min-Max Scaling
   - Unit Vector
   - Feature Extraction
   - PCA

   **Answer:** PCA

   **Explanation:** PCA is suitable for reducing the number of features while retaining the essential information in the data.

### 1. What is the difference between a probability mass function (PMF) and a probability density function (PDF)?

- A PMF is continuous, while a PDF is discrete.
- A PMF is used for continuous random variables, while a PDF is used for discrete random variables.
- A PMF gives the probability of a specific outcome, while a PDF gives the probability density of an outcome.
- A PMF and a PDF are equivalent terms.

**Answer:** A PMF gives the probability of a specific outcome, while a PDF gives the probability density of an outcome.

**Explanation:** The PMF is used for discrete random variables, while the PDF is used for continuous random variables.

### 2. What does the cumulative density function (CDF) represent?

- The probability of a specific outcome.
- The probability density of an outcome.
- The cumulative probability up to a specific value.
- The expected value of a random variable.

**Answer:** The cumulative probability up to a specific value.

**Explanation:** The CDF gives the probability that a random variable takes on a value less than or equal to a specific value.

### 3. Which of the following is an example of a normal distribution?

- The distribution of the number of heads obtained from flipping a fair coin.
- The distribution of the sum of two dice rolls.
- The distribution of heights of adult humans.
- The distribution of the number of cars passing through a toll booth in one hour.

**Answer:** The distribution of heights of adult humans.

**Explanation:** The normal distribution is often used to model the distribution of real-world data, such as the heights of people.

### 4. What is the formula for the cumulative distribution function (CDF)?

- F(x) = P(X > x)
- F(x) = P(X < x)
- F(x) = P(X = x)
- F(x) = P(X >= x)

**Answer:** F(x) = P(X < x)

**Explanation:** The cumulative distribution function (CDF) gives the probability that a random variable X is less than or equal to a certain value x. It is denoted by F(x) and is defined as F(x) = P(X < x).

### 6. What is the Bernoulli distribution?

### 8. What are Z-scores used for?

- To standardize a normal distribution.
- To compute the probability of a specific outcome in a normal distribution.
- To determine if a sample mean is significantly different from a population mean.
- To determine the range of values that a random variable is likely to take on.

**Answer:** To standardize a normal distribution.

**Explanation:** A Z-score is a measure of how many standard deviations a value is from the mean of a normal distribution. It allows for comparison across different normal distributions.
